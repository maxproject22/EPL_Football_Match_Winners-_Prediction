Data Cleaning 

```python
matches["date"] = pd.to_datetime(matches['date'])
```

```python
matches["venue_code"] = matches['venue'].astype('category').cat.codes
matches['opp_code'] = matches['opponent'].astype('category').cat.codes
matches['Hours'] = matches['time'].str.replace(":.+","",regex=True).astype('int')
matches['Winner'] = (matches['result'] == "W").astype('int')
matches["day_code"] = matches['date'].dt.dayofweek
```

Creating Initial ML model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50,min_samples_split=10,random_state=1)
train  = matches[matches['date'] < '2022-01-01']
test  = matches[matches['date'] >= '2022-01-01']
predictor = ['venue_code','opp_code','Hours','day_code']
rf.fit(train[predictor], train['Winner'])
preds = rf.predict(test[predictor])
```

```python
def rolling_Averages(group, col,new_col):
    group = group.sort_values('date')
    rolling_stats = group[col].rolling(3,closed = 'left').mean()
    group[new_col] = rolling_stats
    group = group.dropna(subset=new_col)
    return group
```

```python
col = ['gf','ga','sh','sot','dist','fk','pk','pkatt']

new_col = [f'{c}_rolling' for c in col]
new_col
```

```python
matches_rolling = matches.groupby("team").apply(lambda x : rolling_Averages(x,col,new_col))

matches_rolling = matches_rolling.droplevel('team')
# matches_rolling.index = range(matches_rolling.shape[0])
matches_rolling  = matches_rolling.reset_index(drop=True)
matches_rolling.head(5)
```

```python
matches_rolling[matches_rolling['date'] >= '2022-01-01'].shape
```

Retraining ML Model

```python
def make_prediction(data, predictor):
    train = data[data['date'] < '2022-01-01']
    test = data[data['date'] >= '2022-01-01']
    rf.fit(train[predictor],train['Winner'])
    preds = rf.predict(test[predictor])
    combine = pd.DataFrame(dict(actual=test['Winner'],Predicted=preds),index=test.index)
    precision = precision_score(test['Winner'],preds)
    return combine,precision
```

```python
combine, prediction = make_prediction(matches_rolling,predictor+new_col)
```

```python
class MissingDict(dict):
    __missing__ = lambda self, key: key
    
map_values = {"Brighton and Hove Albion": "Brighton", "Manchester United": "Manchester Utd", "Newcastle United": "Newcastle Utd", "Tottenham Hotspur": "Tottenham", "West Ham United": "West Ham", "Wolverhampton Wanderers": "Wolves"} 
mapping = MissingDict(**map_values)
```

```python
combine['new_team'] = combine['team'].map(mapping)
```

```python
#join to it self to if both team predict winner
merged = combine.merge(combine, left_on=['date','new_team'],right_on=['date','opponent'])
merged.head(2)
```
